## 1. Dependencies/Data Read

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import re
pd.options.display.max_rows = 5000
pd.options.display.max_columns = 100
plt.style.use('ggplot')
```

```python
df = pd.DataFrame(pd.read_csv("Kickstarter.csv"))
for i in range(1,10):
    df = df.append(pd.read_csv("Kickstarter00" + str(i) + ".csv"))
for i in range(11,49):
    df = df.append(pd.read_csv("Kickstarter0" + str(i) + ".csv"))
df = df.reset_index(drop=True)
```

## 3. Data Cleaning 

```python
df = df.drop(['blurb', 'converted_pledged_amount','created_at', 'creator', 'currency_symbol', 'currency_trailing_code', 
              'current_currency', 'id', 'name', 'photo', 'profile', 'slug', 'source_url','state_changed_at', 'static_usd_rate',
              'urls','usd_type','friends', 'is_backing', 'is_starred','permissions'], axis=1)
```

```python
df['subcategory'] = [re.search('(?<="name":")\w+(.[a-zA-Z]+)', df.category.iloc[i]).group() for i in range(0,len(df))]
df['category'] = [re.search('(?<="slug":")\w+', df.category.iloc[i]).group().capitalize()  for i in range(0,len(df))]
df.loc[df['subcategory'] == "Ready-to", 'subcategory'] = "Ready-to-Wear"

df.location=df.location.fillna('short_name":"Undefined')
df['location'] = [re.search('(?<=short_name":")(...)?\s?\w+[a-zA-z]?[-]?\s?\w?\w?[-]?\w?\w?\w?\w?\w?\w?\w?', df.location.iloc[i]).group() for i in range(0,len(df))]
```
